```python
drop_nan=df[df["Row Type"].notna()]
drop_nan
```

```python
column_values=[]
counter=0
for i in drop_nan["Row Type"]:
    if "first name" in i:
        counter+=1
    column_values.append(counter)
```

```python
column_values
```

```python
iter_cols=drop_nan
iter_cols["iteration"]=column_values
iter_cols
```

```python
iter_cols["iteration"]
```

```python
drop_extra_rowna=iter_cols[iter_cols["Row Type"]!="Row Type"]
drop_extra_rowna
```

```python
drop_extra_rowna.head()
```

```python
name_data=drop_extra_rowna[drop_extra_rowna["Row Type"].str.contains("first name")]
name_data
```

```python
name_data.columns
#name_df.drop(columns="")
```

```python
name_data.drop(columns=['Speed1', 'Speed2', 'Electricity',
       'Effort', 'Weight', 'Torque'],inplace=True)
```

```python
name_data
```

```python
name_data.rename(columns={"Row Type":"First Name","Iter Number":"Last Name","Power1":"Date"},inplace=True)
```

```python
name_data
```

```python
name_data["First name"]=name_data["First name"].str[12:]
name_data["Last Name"]=name_data["Last Name"].str[11:]
name_data["Date"]=name_data["Date"].str[6:]
```

```python
name_data
```

```python
Noname_data=drop_extra_rowna[~drop_extra_rowna["Row Type"].str.contains("first name")]  #~ indicator opposite
Noname_data 
```

```python
almost_df=pd.merge(left=name_data,right=Noname_data,how="inner",on="iteration")
```

```python
almost_df.head(
)
```

```python
clean_df=almost_df.drop(columns=["iteration"])
```
